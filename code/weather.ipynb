```python
df_metro = pd.read_csv('Metro_validation.csv')
```

```python
day_of_year = pd.DatetimeIndex(df_metro['VALIDATION_DATE']).dayofweek
df_metro['day_of_year'] = day_of_year
```

```python
metro_data = df_metro.groupby(['day_of_year','lat','long'])['BAND_BOARDINGS_FLOOR'].sum()
```

```python
metro_daily = metro_data.reset_index(name="sum")
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import folium

from folium import plugins
from folium.plugins import HeatMapWithTime


map_hooray = folium.Map(location=[-34.928559, 138.600142],
                        tiles = "CartoDB positron",
                        zoom_start = 14) 

df_hour_list = []

data = metro_daily

max_val = data['sum'].max()
min_val = data['sum'].min()

threshold = data['sum'].mean() + 4* np.sqrt(data['sum'].var())

#data['sum_n'] = (data['sum'] - min_val)/(max_val-min_val)
data['sum_n'] = (data['sum'] > threshold).astype(float)

for day in data.day_of_year.sort_values().unique():
    df_hour_list.append(data.loc[data.day_of_year == day,['long', 'lat', 'sum_n']].values.tolist())

HeatMapWithTime(df_hour_list[::], auto_play=True, display_index=True, index_steps=1,gradient={0.2: 'blue', 0.4: 'lime', 0.6: 'orange', 1: 'red'},radius=10, min_opacity=0, max_opacity=1, use_local_extrema=False).add_to(map_hooray)

map_hooray.save("maps/weekly_validations.html")
map_hooray
```
